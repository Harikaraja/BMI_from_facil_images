```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import cv2
from sklearn import model_selection
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from keras.models import Sequential
```

**Data Set Loaidng**

```python
from google.colab import drive
drive.mount('/content/drive')
path = "/content/drive/MyDrive/Deep Learning projects/BMI.csv"
```

```python
df = pd.read_csv(path,usecols=['Sr.no.', 'BMI']).values
print(df)
```

**Data Preprocessing**

```python
photo_path='/content/drive/MyDrive/Deep Learning projects/photos/'
```

```python
X = []  # the features, or inputs
y = []  # the labels, or outputs


for row in df:
  #print(row[0])
  path = str(int(row[0])) + '.jpg'
  print(path)
  image_bmi = float(row[1])
  image = load_img(photo_path + path, target_size=(128, 128))
  input_arr = img_to_array(image)
  input_arr = np.array(input_arr)
  X.append(input_arr)
  y.append(image_bmi)

```

```python
print(X[1])
print(y)
```

```python
print(y)
```

```python
print(len(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
from tensorflow.keras.layers import Input,Lambda,Dense,Flatten,GlobalAveragePooling2D
from tensorflow.keras.models import Model
from keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
```

```python
IMAGE_SIZE = [128,128]
```

```python
base_model = VGG16(input_shape=IMAGE_SIZE + [3], weights = 'imagenet' ,include_top=False)
```

```python
# Add Global Average Pooling layer
x = GlobalAveragePooling2D()(base_model.output)
```

```python
# Add a custom regression output layer
regression_layer = Dense(1, activation='linear')(x)
```

```python
for layer in base_model.layers:            # retraining of weights is not necessary because we are using state of art model
  layer.trainable = False
```

```python
# Create the new model
model = Model(inputs=base_model.input, outputs=regression_layer)
```

```python
model.summary()
```

```python
# Compile the model with an appropriate regression loss function (e.g., Mean Squared Error)
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])
```

```python
epochs = 50
batch_size = 32
history = model.fit(X_train,y_train,epochs=epochs, batch_size=batch_size)
```

```python
# Save the model
model.save('/content/drive/MyDrive/Deep Learning projects/model.h5')
```
